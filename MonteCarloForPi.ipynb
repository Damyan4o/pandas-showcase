```python
#define a function
# for loop in 100 000
# 10/100/1000/10000 print function
import random
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as pltp
def monte_carlo():
    count = 0
    A = np.zeros((1000000,2),dtype = float)
    for i in range(1000000):
        x = random.uniform(-5,5)
        A[i,0]= x
        y = random.uniform(-5,5)
        A[i,1] = y
        if x**2+y**2 <= 5:
            count+=1
        if i+1 == 10:
            print('After 10 runs:',count*20/(i+1))
        if i+1 == 100:
            print('After 100 runs:',count*20/(i+1))
        if i+1 == 1000:
            print('After 1000 runs:',count*20/(i+1))
        if i+1 == 10000:
            print('After 10000 runs:',count*20/(i+1))
        if i+1 == 100000:
            print('After 100000 runs:',count*20/(i+1))
        if i+1 == 1000000:
            print('After 1000000 runs:',count*20/(i+1))
    B = pd.DataFrame(data = A, columns = ['x','y'])
    B1 = B.iloc[0:999]
    B1.plot(kind = 'scatter', x = 'x', y = 'y', s = 0.1, xticks = np.arange(-5,6,2), yticks = np.arange(-5,6,2))
    a = 5*np.cos(np.linspace(0, 2 * np.pi, 200))
    b = 5*np.sin(np.linspace(0, 2 * np.pi, 200))
    pltp.plot(a, b, color="red")
    pltp.gca().set_aspect('equal')
    pltp.show()
monte_carlo()
```
